```python
from sklearn.datasets import load_iris
iris=load_iris()
x=iris.data
y=iris.target
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy')
model=model.fit(x,y)
op=model.predict([[5.3,9.3,7.2,5.2]])
print(op)
```

```python
fn=iris.feature_names
cn=iris.target_names
import graphviz
op=tree.export_graphviz(model,feature_names=fn,class_names=cn,filled=True,rounded=True,special_characters=True)
graph=graphviz.Source(op)
graph
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2529)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=DecisionTreeClassifier(criterion='entropy')
model=model.fit(x_train,y_train)
```

```python
op=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,op))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
print("confsion_matrix")
cm=confusion_matrix(y_test,op)
print(cm)
sns.heatmap(cm,annot=True,cmap="Reds")
plt.title("confusion matrix")
plt.xlabel("predicted")
plt.ylabel("actual")
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,op))
```
