```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data={"Hours":[1,2,3,4,5,6],
      'marks':[50,60,70,80,90,100]}
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
x = df[['Hours']].values
y = df['marks'].values
```

```python
x
```

```python
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
print(f"slope: {slope}")
print(f"intercept: {intercept}")
print(f"equation: y = {slope}x + {intercept}")
```

```python
df["predicted marks"] = model.predict(x)
```

```python
df
```

```python
plt.scatter(df['Hours'],df['marks'],color="blue",label="actual marks")
```

```python
plt.plot(df['Hours'],df['marks'],color="pink",label="actual marks")
plt.xlabel('Hours')
plt.ylabel('Marks')
plt.title('LinearRegression')
plt.legend()
plt.show()
```

```python
#encoding using numbers
```

```python
file.replace({"fuel":{"Diesel":0,"Petrol":1,"CNG":2,"LPG":3}},inplace=True)
file.replace({"seller_type":{"Individual":0,"Dealer":1,"Trustmark Dealer":2}},inplace=True)
file.replace({"transmission":{"Manual":0,"Automatic":1}},inplace=True)
file.replace({"owner":{"First Owner":0,"Second Owner":1,"Third Owner":2,"Fourth & Above Owner":3,"Test Drive Car":4}},inplace=True)
```

```python
y=file["selling_price"]
```

```python
y.shape
```

```python
y
```

```python
x=file[["year","km_driven","fuel","seller_type","transmission","owner"]]
```

```python
x
```

```python
#get train test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2529)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("actual prices")
plt.ylabel("predicted prices")
plt.title("Actual price vs Predicted price")
plt.show()
```
