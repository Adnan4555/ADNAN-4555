importing *images*

```python
import matplotlib.pyplot as plt
```

```python
random_image = np.random.randint(1,255,size=(225,225,3))
```

```python
plt.imshow(random_image)
plt.show()
```

```python
color = np.array([255,0,-100]).reshape(1,1,3)
```

```python
image = color + random_image
image_clip = np.clip(image,0,255)
```

```python
plt.imshow(image_clip)
plt.show()
```

```python
v = np.vstack((image_clip,random_image))
```

```python
plt.imshow(v)
plt.show()
```

```python
v = np.hstack((image_clip,random_image))
```

```python
plt.imshow(v)
plt.show()
```
