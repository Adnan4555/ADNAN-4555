```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,LSTM
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("/content/IMDB Dataset.csv")
```

```python
df.head()
```

```python
df["sentiment"].value_counts()
```

```python
df['sentiment']= LabelEncoder().fit_transform(df['sentiment'])
```

```python
df.head()
```

```python
train_df,test_df = train_test_split(df,test_size=0.2,random_state=2529)
```

```python
print(train_df.shape)
```

#data preprocessing

```python
#tokenize the data
tokenizer = Tokenizer(num_words=5000) #creating instance of the tokeniszer
#tokenizer are used to convert words into integers,num words will convert most frequent 5000 words
tokenizer.fit_on_texts(train_df["review"])
x_train = pad_sequences(tokenizer.texts_to_sequences(train_df["review"]),maxlen=200) #make sure that data has same length
x_test = pad_sequences(tokenizer.texts_to_sequences(test_df["review"]),maxlen=200)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
y_train = train_df["sentiment"]
y_test = test_df ["sentiment"]
```

```python
#build the model

model = Sequential()
model.add(Embedding(input_dim=5000,output_dim=128,input_length=200))
model.add(LSTM(128,dropout=0.2,recurrent_dropout=0.2))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=5,batch_size=64,validation_split=0.2)
```

#building a predictive system

```python
def predict_sentiment(review):
  #tokenize and pad the review
  sequence = tokenizer.texts_to_sequences([review])
  padded_sequence = pad_sequences(sequence,maxlen=200)
  prediction = model.predict(padded_sequence)
  sentiment = "positive" if prediction[0][0] > 0.5 else"negative"
  return sentiment
```

```python
#example usage
new_review = "this movie was fantastic. i loved it."
sentiment = predict_sentiment(new_review)
print(f"the sentiment of the review is: {sentiment}")
```
