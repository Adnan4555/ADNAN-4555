```python
#cmap
x1 = [1,2,3,4,5]
y1 = [10,24,36,45,50]
color=["red","green","blue","yellow","pink"]
sizes=[100,250,300,340,380]
plt.scatter(x1,y1,c=color,s=sizes,cmap="viridis")
plt.colorbar()
plt.show()
```

```python
#vertical bar plot
x=[1,2,3,4,5]
y=[10,24,36,45,50]
plt.bar(x,y,color="green",width=0.5)
plt.show()
```

```python
#horizontal bar plot
x=[1,2,3,4,5]
y=[10,24,36,45,50]
plt.barh(x,y,color="orange",height=0.5)
plt.show()
```

```python
#vertical stacked bar graph
years = [2018,2019,2020,2021]
placements = [150,250,270,430]
bca = [67,87,78,90]
cs = [78,90,87,67]
bcom = [65,67,87,79]
plt.title("PLACEMENTS HUB")
plt.bar(years,bca,color="green",width=0.5)
plt.bar(years,cs,bottom=bca,color="blue",width=0.5)
plt.bar(years,bcom,color="orange",width=0.5)
plt.show()
```

```python
#horizonatal stacked bar graph
years = [2018,2019,2020,2021]
placements = [150,250,270,430]
bca = [67,87,78,90]
cs = [78,90,87,67]
bcom = [65,67,87,79]
plt.title("PLACEMENTS HUB")
plt.barh(years,bca,color="green",height=0.5)
plt.barh(years,cs,left=bca,color="blue",height=0.5)
plt.barh(years,bcom,color="orange",height=0.5)
plt.show()
```

```python
#piechart
labels = ["AI projects","web development","Mobile apps","data science"]
sizes = [15,30,45,10]
plt.pie(sizes,labels=labels,startangle=90)
plt.axis("equal")
plt.title("MY PROJECTS")
plt.savefig("pie_chart.png")
plt.show()
```

```python
import matplotlib.image as img

image = img.imread("/content/bently.jpeg")
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
image.shape
```

```python
#HORIZONTAL FLIP
flip = np.fliplr(image)
plt.imshow(flip)
plt.axis("off")
plt.show()
```

```python
#VERTICAL FLIP
flip = np.flipud(image)
plt.imshow(flip)
plt.axis("off")
plt.show()
```

```python
#ROTATE
rotate = np.rot90(image)
plt.imshow(rotate)
plt.axis("off")
plt.show()
```

```python
#ROTATE ARBITARY ANGLE
from scipy.ndimage import rotate
rotate = rotate(image,45,reshape=True)
plt.imshow(rotate)
plt.axis("off")
plt.show()
```
